```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrada: nombre, calorías, proteína, grasa, carbohidratos de los alimentos
# Este es un ejemplo de cómo podrían organizarse los datos. Puedes cargar los datos desde un archivo CSV u otra fuente.

objetivo_grasa = 50  # Gramos de grasa objetivo por día
objetivo_carbohidratos = 200  # Gramos de carbohidratos objetivo por día
calorias_maximas = 2500  # Calorías máximas por día
generaciones = 100  # Número de generaciones
tamanio_poblacion = 100  # Tamaño de la población
prob_mutacion = 0.01  # Probabilidad de mutación

def evaluar_dieta(dieta):
    total_calorias = 0
    total_proteina = 0
    total_grasa = 0
    total_carbohidratos = 0

    for i, cantidad in enumerate(dieta):
        alimento = alimentos[i]
        total_calorias += cantidad * alimento["calorias"]
        total_proteina += cantidad * alimento["proteina"]
        total_grasa += cantidad * alimento["grasa"]
        total_carbohidratos += cantidad * alimento["carbohidratos"]

    penalizacion = 0  # Puedes agregar penalizaciones si se exceden ciertos límites

    if total_grasa > objetivo_grasa:
        penalizacion += (total_grasa - objetivo_grasa) * 10  # Penalización por exceso de grasa

    if total_carbohidratos > objetivo_carbohidratos:
        penalizacion += (total_carbohidratos - objetivo_carbohidratos) * 5  # Penalización por exceso de carbohidratos

    if total_calorias > calorias_maximas:
        penalizacion += (total_calorias - calorias_maximas)  # Penalización por exceso de calorías

    aptitud = 1 / (1 + penalizacion)  # Cuanto mayor es la aptitud, mejor la dieta

    return aptitud

def generar_dieta(tamanio_poblacion):
    poblacion = []
    for _ in range(tamanio_poblacion):
        dieta = [random.randint(0, 10) for _ in alimentos]  # Generar una dieta aleatoria
        poblacion.append(dieta)
    return poblacion

def cruzar_padres(padre1, padre2):
    punto_cruce = random.randint(1, len(padre1) - 1)
    hijo1 = padre1[:punto_cruce] + padre2[punto_cruce:]
    hijo2 = padre2[:punto_cruce] + padre1[punto_cruce:]
    return hijo1, hijo2

def mutar_dieta(dieta, prob_mutacion):
    for i in range(len(dieta)):
        if random.random() < prob_mutacion:
            dieta[i] = random.randint(0, 10)  # Cambiar la cantidad de un alimento de forma aleatoria
    return dieta

# Inicialización de la población
poblacion = generar_dieta(tamanio_poblacion)
mejor_dieta = None
mejor_aptitud = 0

# Evolución de la población
for generacion in range(generaciones):
    # Evaluar la aptitud de la población
    aptitudes = [evaluar_dieta(dieta) for dieta in poblacion]

    # Encontrar la mejor dieta de esta generación
    mejor_indice = np.argmax(aptitudes)
    if aptitudes[mejor_indice] > mejor_aptitud:
        mejor_dieta = poblacion[mejor_indice]
        mejor_aptitud = aptitudes[mejor_indice]

    # Selección de padres
    padres_seleccionados = np.random.choice(tamanio_poblacion, size=tamanio_poblacion, p=aptitudes/np.sum(aptitudes))

    # Crear la nueva generación mediante cruce y mutación
    nueva_poblacion = []
    for i in range(0, tamanio_poblacion, 2):
        padre1 = poblacion[padres_seleccionados[i]]
        padre2 = poblacion[padres_seleccionados[i + 1]]
        hijo1, hijo2 = cruzar_padres(padre1, padre2)
        hijo1 = mutar_dieta(hijo1, prob_mutacion)
        hijo2 = mutar_dieta(hijo2, prob_mutacion)
        nueva_poblacion.extend([hijo1, hijo2])

    poblacion = nueva_poblacion

    print(f"Generación {generacion}: Mejor aptitud = {mejor_aptitud}")

# Resultado final
print("Mejor dieta encontrada:")
for i, cantidad in enumerate(mejor_dieta):
    if(cantidad != 0):
        print(f"{alimentos[i]['nombre']}: {cantidad} unidades")

# Puedes agregar visualización o análisis adicional según tus necesidades
```
